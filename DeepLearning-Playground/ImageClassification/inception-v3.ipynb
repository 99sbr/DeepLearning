```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.python.keras.callbacks import TensorBoard

df = pd.read_csv("../input/snakesinhood/dataset/train.csv")
NB_CLASSES = len(df.breed.value_counts())
df.image_id = df.image_id.apply(lambda x: x + '.jpg')
train_df, val_df = train_test_split(df, random_state=24, stratify=df['breed'], test_size=0.2)

# Configs
# IMG_SIZE is determined by EfficientNet model choice
IM_WIDTH, IM_HEIGHT = 299 , 299

TF_EPOCH = 100
FT_EPOCH = 100
NB_IV3_LAYERS_TO_FREEZE = 172


def add_callback(name):
    weightsOutputFile = name+'_'+ 'Inception_v3.{epoch:02d}-{val_accuracy:.2f}.hdf5'
#     tensor_board = TensorBoard(log_dir='Graph', histogram_freq=0, batch_size=32, write_graph=True, write_images=True)
    learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience=5, verbose=1, factor=0.5,
                                                min_lr=0.0000001)
    checkpoint = ModelCheckpoint(weightsOutputFile, monitor='val_accuracy', verbose=1, save_best_only=True,
                                 save_weights_only=False,
                                 mode='auto', period=1)
    # early = EarlyStopping(monitor='val_acc', min_delta=0, patience=10, verbose=1, mode='auto')
    if name=="tl":
        return learning_rate_reduction
    return [checkpoint, learning_rate_reduction]


class Generators:
    """
    Train, validation and test generators
    """

    def __init__(self, train_df, val_df):
        self.batch_size = 128
        self.img_size = (IM_WIDTH, IM_HEIGHT)

        # Base train/validation generator
        train_datagen = ImageDataGenerator(
            rescale=1. / 255.,
            rotation_range=90,
            width_shift_range=0.2,
            height_shift_range=0.2,
            zoom_range=0.4,
            fill_mode='nearest',
            horizontal_flip=True,
            vertical_flip=True
        )
        
        val_datagen = ImageDataGenerator(
            rescale=1. / 255.,
            fill_mode='nearest',
            horizontal_flip=True,
            vertical_flip=True
        )

        # # Validation generator
        self.val_generator = val_datagen.flow_from_dataframe(
            dataframe=val_df,
            directory="../input/snakesinhood/dataset/train/",
            x_col="image_id",
            y_col="breed",
            batch_size=self.batch_size,
            seed=42,
            shuffle=False,
            class_mode="categorical", color_mode='rgb', interpolation='bicubic',
            target_size=self.img_size)
        print('Validation generator created')

        # Train generator
        self.train_generator = train_datagen.flow_from_dataframe(
            dataframe=train_df,
            directory="../input/snakesinhood/dataset/train/",
            x_col="image_id",
            y_col="breed",
            batch_size=self.batch_size,
            seed=42,
            shuffle=False,
            class_mode="categorical", color_mode='rgb', interpolation='bicubic',
            target_size=self.img_size)

        print('Train generator created')


# Create generators
generators = Generators(train_df, val_df)
print("Generators created")


def setup_to_transfer_learn(num_classes):
    inputs = layers.Input(shape=(IM_WIDTH, IM_HEIGHT, 3))
    from keras.applications.inception_v3 import InceptionV3
    model = InceptionV3(include_top=False, input_tensor=inputs, weights='imagenet')

    # Freeze the pretrained weights
    model.trainable = False

    # Rebuild top
    x = layers.GlobalAveragePooling2D(name="avg_pool")(model.output)
    x = layers.BatchNormalization()(x)

    top_dropout_rate = 0.2
    x = layers.Dropout(top_dropout_rate)(x)
    x = layers.Dense(512, activation="relu")(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    # Compile
    model = tf.keras.Model(inputs, outputs)

    optimizer = tf.keras.optimizers.Nadam(learning_rate=1e-2)
    model.compile(
        optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"]
    )
    return model


def setup_to_finetune(model):
    for layer in model.layers[:NB_IV3_LAYERS_TO_FREEZE]:
        layer.trainable = False
    for layer in model.layers[NB_IV3_LAYERS_TO_FREEZE:]:
        if not isinstance(layer, layers.BatchNormalization):
            layer.trainable = True

    optimizer = tf.keras.optimizers.Nadam(learning_rate=1e-4)
    model.compile(
        optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"]
    )

```

```python
model = setup_to_transfer_learn(num_classes=NB_CLASSES)
history1 = model.fit_generator(
    generators.train_generator,
    steps_per_epoch=generators.train_generator.n // generators.train_generator.batch_size,
    epochs=TF_EPOCH,
    shuffle=False,
    verbose=2,
    validation_data = generators.val_generator,
    validation_steps = generators.val_generator.n // generators.val_generator.batch_size,
    callbacks=add_callback('tl'))
```

```python
setup_to_finetune(model)
```

```python
history2 = model.fit_generator(
    generators.train_generator,
    steps_per_epoch=generators.train_generator.n // generators.train_generator.batch_size,
    epochs=FT_EPOCH,
    shuffle=False,
    verbose=2,
    validation_data = generators.val_generator,
    validation_steps = generators.val_generator.n // generators.val_generator.batch_size,
    callbacks=add_callback('ft'))
```
